### - Fonction d'imputation des Nan par la médiane en fonction de la marge basse

```python
def imputation_Nan_Par_la_mediane(colonnes):
    margin_low = colonnes [0]
    is_genuine = colonnes [1]
    if pd.isnull(margin_low):
        if is_genuine == 4.11:
            return True
        else:
            return False
    else:
        return margin_low
df.head(10)
```

```python
df['margin_low'] = df[['margin_low', 'is_genuine']].apply(imputation_Nan_Par_la_mediane, axis=1)
```

## 2. K-means

```python
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_train_scaled)
```

* Mappage des cluster : ils ne correspondent pas aux classes

* Trouvons quel cluster correspond à quelle classe

```python
from sklearn.metrics import confusion_matrix

y_pred_kmeans = kmeans.predict(X_test_scaled)
conf_mat = confusion_matrix(y_test, y_pred_kmeans)
```

```python
# Si la diagonale principale n'est pas la plus grande, inverser les clusters
if conf_mat[0,0] + conf_mat[1,1] < conf_mat[0,1] + conf_mat[1,0]:
    y_pred_kmeans = 1 - y_pred_kmeans
```

**- Matrice de confusion**

```python
print(confusion_matrix(y_test, y_pred_kmeans))
```

**- Classification Report**

```python
print(classification_report(y_test, y_pred_kmeans))
```

**- Accuracy**

```python
print(f"Accuracy: {accuracy_score(y_test, y_pred_kmeans):.5f}")
```
