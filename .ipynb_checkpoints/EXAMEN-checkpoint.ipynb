### Fonction d'imputation des Nan par la médiane en fonction de la marge basse

```python
def imputation_Nan_Par_la_mediane(colonnes):
    margin_low = colonnes [0]
    is_genuine = colonnes [1]
    if pd.isnull(margin_low):
        if is_genuine == 4.11:
            return True
        else:
            return False
    else:
        return margin_low
    

df.head(10)
```

### Appel de la fonction

```python
df['margin_low'] = df[['margin_low', 'is_genuine']].apply(imputation_Nan_Par_la_mediane, axis=1)
```

<h2 style="margin: auto; padding: 20px 0; color:#fff; text-align: center">K MEANS</h2>
</div> 

**1. Standardisation des données**


```python
from sklearn.preprocessing import StandardScaler
# déclaration des variables interressantes pour le clustering
var = ['length', 'height_left', 'height_right', 'margin_low', 'margin_up', 'diagonal']
X = df[var]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

**2. Détection du nombre de cluster**

```python
# Détection automatique du coude avec Yellowbrick
model = KMeans(random_state=42)
visualizer = KElbowVisualizer(model, k=(2,10), timings = False)

visualizer.fit(X_scaled)
visualizer.show()
```

**3. Construction du modele segmentation avec l'algo des kmeans ()**

**Standardisation**

```python
# K-Means avec le bon k
features = ['length', 'height_left', 'height_right', 'margin_low', 'margin_up', 'diagonal']
X = df[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

**4. Evaluons le modèle kmeans avec la silhouette**

```python
silhouette_scores = []
k_range = range(2, 10)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    silhouette_scores.append(score)

best_k = k_range[silhouette_scores.index(max(silhouette_scores))]
print(f"Meilleur K basé sur le score de silhouette : {best_k}")

kmeans_final = KMeans(n_clusters=best_k, random_state=42)
df['cluster'] = kmeans_final.fit_predict(X_scaled)
```

**5. Visualisons notre cluster**

```python
# Gestion des valeurs manquantes (remplacer par la médiane)
imputer = SimpleImputer(strategy='median')
X_imputed = imputer.fit_transform(X)

# Standardisation des données (important pour K-means)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# Application de K-means avec K=2
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_scaled)

# Ajout des labels au DataFrame original
df['cluster'] = kmeans.labels_

# Visualisation des clusters (en 2D pour exemple)
plt.figure(figsize=(10, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_, cmap=plt.cm.Blues)
plt.title('Clusters des billets (K=2)')
plt.xlabel('diagonal (standardisée)')
plt.ylabel('height_left (standardisée)')
plt.colorbar(label='Cluster')
plt.show()

# Statistiques descriptives par cluster
print(df.groupby('cluster')[var].mean())

# Comparaison avec la vérité terrain (si disponible)
if 'is_genuine' in df.columns:
    from sklearn.metrics import confusion_matrix
    cm = confusion_matrix(df['is_genuine'], df['cluster'])
    print("\nMatrice de confusion vs vraie étiquette:")
    print(cm)
```

**Profiling**

```python
df['cluster'] = df['cluster'].astype(str)
profiling = df.groupby('cluster')[features].mean()
display(profiling)
```

<h2 style="margin: auto; padding: 20px 0; color:#fff; text-align: center">KNN</h2>
</div> 

**1. Supprimons les valeurs manquantes**

```python
df_clean = df.dropna().copy()
```

**2. Séparons les variables et les target**

```python
X = df_clean[['diagonal', 'height_left', 'height_right', 'margin_low', 'margin_up', 'length']]
y = df_clean['is_genuine']
```

**3. Convertissons les booléens en int (True->1, False->0)**

```python
y = y.astype(int)
```

**4. Recherchons le meilleur K**

- grille de paramètres à tester

```python
param_grid = {'n_neighbors': np.arange(1, 30)}
```

* Initialisation du GridSearchCV

```python
knn = KNeighborsClassifier()
knn_cv = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')
knn_cv.fit(x_train_scaled, y_train)
```

* Meilleurs paramètres

```python
best_k = knn_cv.best_params_['n_neighbors']
print(f"Meilleur K trouvé: {best_k}")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(param_grid['n_neighbors'], knn_cv.cv_results_['mean_test_score'], marker='o')
plt.xlabel('Nombre de voisins (K)')
plt.ylabel('Accuracy moyenne')
plt.title('Performance du KNN en fonction de K')
plt.axvline(best_k, color='red', linestyle='--')
plt.show()
```

**4. Entraînons le modèle final**

* Initialisons avec le meilleur K

```python
final_knn = KNeighborsClassifier(n_neighbors=best_k)
```

* Entraînement du modèle

```python
final_knn.fit(x_train_scaled, y_train)
```

**5. Évaluons notre modèle**

```python
y_pred = final_knn.predict(x_test_scaled)
y_prob = final_knn.predict_proba(x_test_scaled)[:, 1]  # Probabilités pour la classe positive
```

**6. Métriques d'évaluation**

```python
print("\nPerformance sur l'ensemble de test:")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2%}")
print(f"AUC ROC: {roc_auc_score(y_test, y_prob):.2%}\n")
```

**7. Matrice de confusion**

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Faux', 'Vrai'], 
            yticklabels=['Faux', 'Vrai'])
plt.xlabel('Prédit')
plt.ylabel('Réel')
plt.title('Matrice de confusion')
plt.show()
```

**8. Rapport de classification**

```python
print("\nRapport de classification détaillé:")
print(classification_report(y_test, y_pred, target_names=['Faux', 'Vrai']))
```

**9. Courbe ROC**

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'KNN (AUC = {roc_auc_score(y_test, y_prob):.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('Taux de faux positifs')
plt.ylabel('Taux de vrais positifs')
plt.title('Courbe ROC')
plt.legend()
plt.show()
```
